```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
from sklearn.model_selection import GridSearchCV
```

# predict if a person would be satisfied with flight 
data -> `https://www.kaggle.com/datasets/teejmahal20/airline-passenger-satisfaction`

```python
data= pd.read_csv('flight_L.csv')
data
```

```python
for i in data['Class']:
    if(i=='Business'):
        data['Class'].replace(i, 1, inplace=True)
    elif(i=='Eco'):
        data['Class'].replace(i, 2, inplace=True)
    elif(i=='Eco Plus'):
        data['Class'].replace(i, 3, inplace=True)
        
for i in data['Customer Type']:
    if(i=='Loyal Customer'):
        data['Customer Type'].replace(i, 1, inplace=True)
    elif(i=='disloyal Customer'):
        data['Customer Type'].replace(i, 0, inplace=True)

for i in data['Type of Travel']:
    if(i=='Personal Travel'):
        data['Type of Travel'].replace(i, 0, inplace=True)
    elif(i=='Business travel'):
        data['Type of Travel'].replace(i, 1, inplace=True)
        
for i in data['Gender']:
    if(i=='Male'):
        data['Gender'].replace(i, 1, inplace=True)
    elif(i=='Female'):
        data['Gender'].replace(i, 2, inplace=True)
        
for i in data['satisfaction']:
    if(i=='neutral or dissatisfied'):
        data['satisfaction'].replace(i, 0, inplace=True)
    elif(i=='satisfied'):
        data['satisfaction'].replace(i, 1, inplace=True)
```

```python
data.drop(columns=['Unnamed: 0','id'],inplace=True)
data.dropna(inplace=True)
data
```

```python
X=data.drop(columns=['satisfaction'])
y=data['satisfaction']
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=.2)
params = {'n_estimators': [50, 100,500,1000],
                'max_depth': [None,10],
                'min_samples_split': [2,8],
                'min_samples_leaf': [1],
                'max_features': ['log2','sqrt'],}

gs_clf = GridSearchCV(estimator=RandomForestClassifier(n_jobs=-1),param_grid=params,cv=5,verbose=2,n_jobs=-1)

gs_clf.fit(x_train,y_train)
y_preds =gs_clf.predict(x_test)
gs_clf.score(x_test,y_test)

```
