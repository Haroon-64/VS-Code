```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as pt
```

labeling

```python
x= (1,2,3,4,5)
y= (2,4,8,16,32)
pt.title('pla')
pt.xlabel('x-axis')
pt.ylabel('y-axis')
```

```python
x_x=(1,2,3,4,5)
x_y=(1,4,9,16,25)
pt.plot(x_x,x_y)
pt.plot(x,y)   #stateless ploting  / pylpot api-> less flexible
pt.legend(['bit','sq'])
pt.show()
```

```python
x_a=(1,2,3,4,5)
x_b=(1,4,9,16,25)
pt.plot(x_a,x_b,label='another way')
pt.plot([1,2,3],[1,2,3],label='to legend')
pt.legend()
pt.show()
```

```python
x=np.random.randn(10000)
figg, axs = pt.subplots( figsize=(10, 4))
axs.hist(x,bins=1000);
```

```python
figg ,((ax1, ax2), (ax3, ax4)) = pt.subplots(nrows=2,ncols=2,figsize= (10, 10))
ax1.plot(x, x**2, x, x**3)
ax2.plot(x, np.sin(x), x, np.cos(x))
ax3.plot(x, np.exp(x))
ax4.plot(x, x) 
```

```python
fig,ax=pt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,np.sin(x))
ax[0,1].plot(x,np.cos(x))
ax[1,0].plot(x,np.tan(x))
ax[1,1].plot(x,np.exp(x))
```
